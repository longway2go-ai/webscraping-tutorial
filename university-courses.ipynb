```python
rows = []

for j in range(1, 11):
    url = f'https://www.tum.de/en/studies/degree-programs#page={j}'
    webpage = requests.get(url).text
    soup = BeautifulSoup(webpage, 'lxml')

    names = [i.text.strip() for i in soup.find_all('p', class_='roofline')]
    degrees = [i.text.strip() for i in soup.find_all('h3')]
    intros = [i.text.strip() for i in soup.select('div.list-teaser__text > p')]

    for name, degree, intro in zip(names, degrees, intros):
        rows.append({'name': name, 'degree': degree, 'intro': intro})

final = pd.DataFrame(rows)
```
